# Problem 1:  The conjugate and the dual
Consider the "monotropic" program
\begin{align}
\text{minimize} & \quad  \|x\|_\infty \\
\text{subject to} & \quad Ax=b . \nonumber
\end{align}

Write this as an unconstrained (or implicitly constrained) problem using the characteristic function of the zero vector $\chi_0(z) .$  This function is zero if its argument is zero, and infinite otherwise.



Writing the problem in Saddle-point form:
$$ \min_{x} \max_{\lambda} \|x\|_{\infty} + \langle \lambda, Ax-b \rangle $$
Using the characteristic function $\chi_{0}(Ax+b)$, we can write the objective as
$$ \text{minimize } \|x\|_{\infty} + \chi_{0}(Ax+b) $$
where $ \chi_{0}(z) = \begin{cases}
0, \text{ if } z=0\\
\infty , \text{ otherwise}\\
\end{cases} $

#### a) What is the conjugate of $f(z)= \|z\|_\infty$?

$$ f^{*}(y) = \chi_{1}(y) = \begin{cases}
0, \text{ if } |z|\le1\\
\infty , \text{ if } |z|\ge1\\
\end{cases} $$

#### b) What is the conjugate of $g(z)=\chi_0(z)$?

The conjugate of a function:
$$ g^{*}(y) = \max_{z} y^{T}z - g(z) $$
$$ \Rightarrow g^{*}(y) = \max_{z} y^{T}z - \chi_{0}(z) $$
Although, this function is stable (doesn't blow up) at $z=0$ and its value there should be equal to $0$.

#### c) Using the conjugate functions, write down the dual of the monotropic problem.

We start with the definition of the dual function,
$$ d(\lambda) = \min_{x}f(x) + \langle \lambda, Ax-b\rangle $$
$$ = \min_{x} f(x) + \lambda^{T}Ax - \lambda^{T}b $$
$$ = \min_{x} \|x\|_{\infty} + \lambda^{T}Ax - \lambda^{T}b $$
$$ = \lambda^{T}b - \max_{x}(-\|x\|_{\infty} - \lambda^{T}Ax) $$
$$ = \langle \lambda, b \rangle - f^{*}(-A^{T}\lambda) $$
We know that $ f^{*}(y) = \chi_{1}(y) $. Applying this knowledge,
$$ d(\lambda) = \langle \lambda, b \rangle - \chi_{1}(-A^{T}\lambda).$$



# Problem 2:  Linear programming
Consider the linear program
\begin{align*}
\text{minimize} \quad &  c^T x \\
\text{subject to} \quad & Ax=b\\
 & x\ge 0.
\end{align*}

#### a) Write the optimality conditions for this problem (i.e., the KKT system).  Your solution should contain 5 equations.

Let $f(x) = c^{T}x$, $h(x) = Ax-b$ and $g(x) = -x$. Optimality conditions:
1. $ c + A^{T}\lambda - \eta = 0 $
2. $-x^{*} \le 0$
3. $Ax^{*}-b = 0$
4. $\eta \ge 0$
5. $-\eta_{i}x_{i}^{*} = 0$


#### b) Write the Lagrangian for this problem.
Use $\lambda$ and $\eta$ are your lagrange multipliers for the equality and inequality constraints, respectively.

$$L(x, \lambda, \eta) = c^{T}x + \langle \lambda, Ax-b \rangle + \langle \eta, -x \rangle$$

#### c) Minimize out the primal variables in the Lagrangian, and write the dual formulation of this linear program.

We need to optimize for $x$ first. We know that the dual function to the optimization problem is
$$ d(\lambda, \eta) = \min_{x} L(x, \lambda, \eta) $$
$$ \text{minimize } c^{T}x + \langle \lambda, Ax-b \rangle + \langle \eta, -x \rangle $$
$$ = \min_{x} c^{T}x + \lambda^{T}(Ax-b) - \eta^{T}x $$
$$ = \min_{x}(c^{T} + \lambda^{T}A - \eta^{T})x - \lambda^{T}b $$
From 1 we take that $ c + A^{T}\lambda - \eta = 0 $
$$ \Rightarrow d(\lambda, \eta) = - \lambda^{T}b $$



# Problem 3: Sensitivity bound
Consider the problem
\begin{align}
\text{minimize} \quad &  f(x) \\
\text{subject to}\quad & g(x) \le 0.
\end{align}
Let $x_0$ be a solution to this problem, and $\lambda_0$ be the corresponding optimal Lagrange multiplier. Now, define a perturbed problem
  \begin{align}
\text{minimize}\quad &  f(x) \\
\text{subject to}\quad & g(x) \le \epsilon
\end{align}
where $\epsilon$ is a vector. Let $x_\epsilon$ be a solution to the perturbed problem.  Note, if we put large negative values in $\epsilon$, then  the constraint set gets smaller, and we expect the corresponding value of $f(x_\epsilon)$ to increase.  

Prove the "sensitivity bound"
$$ f(x_0) -   \lambda_0^T \epsilon \le f(x_\epsilon). $$
This bound shows that the Lagrange multipliers determine how much the objective increases as the vector $\epsilon$ becomes more negative.

Let $x_{0}$ be the solution to the primal problem and $\lambda_{0}$ be the solution to the dual problem. Assuming strong duality, we have
$$ f(x^{*}) = \max_{\lambda, \eta} d(\lambda, \eta) $$
From the definition of the dual function, we have
$$ d(\lambda, \eta) \le f(x) + \lambda_{0}^{T}g(x) $$
We know that $x_{\epsilon}$ is the solution to the perturbed problem and $g(x) \le \epsilon$. Applying this knowledge,
$$ \Rightarrow f(x_{0}) \le f(x_{\epsilon}) + \lambda_{0}^{T}\epsilon $$
$$ \Rightarrow f(x_{0}) - \lambda_{0}^{T}\epsilon \le f(x_{\epsilon}). $$
