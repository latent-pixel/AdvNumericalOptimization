### (a) Consider the function
 $$f(x) = \begin{cases}
0, \text{ if } x<0\\
\infty , \text{ if } x\ge 0\\
\end{cases}.$$

Is this function closed and proper?

- The given function is *closed* since the epigraph $ \{(x, y) | f(x) \le y \}$ is closed, i.e, the function $f$ has a value for each value of $x$.

- A function can be called *proper* if the epigraph is non-empty. We know that 
$$ epi(f) = \{(x, y) | f(x) \le y \}. $$
This means that the function $f(x)$ must have a bound $y$. Here, as the value of the $x$ increases to a positive value, $f(x)$ becomes unbounded. Hence, the function is not proper.

### (b) Consider the set of $n\times n$ low-rank matrices $\{A\in\mathbb{R}^{n\times n} | \, \text{rank}(A)\le k \},$ where $0<k<n$.   Is this set convex?  Why?

A set is convex if all convex combinations lie inside the set. Here, we have a set of $n \times n$ low-rank matrices - consider the following two matrices:

 $$\{A_1 \in \mathbb{R}^{n \times n} | \, \text{rank}(A_1)\le k \}$$
 $$\{A_2 \in \mathbb{R}^{n \times n} | \, \text{rank}(A_2)\le k \}$$

where $0 < k < n$. Using the subadditivity property of the rank of a matrix,

$$ rank(\theta A_1 + (1 - \theta) A_2) \le rank(A_1) + rank(A_2) - (\theta \,\,is \,\, a \,\, scalar)$$
$$ rank(\theta A_1 + (1 - \theta) A_2) \le 2k $$

We can see that the linear combination of the two matrices doesn't necessarily have the same rank as that of the two original matrices. Therefore, the set is non-convex.

### (c) Suppose $\{A_i\}$ are matrices of identical size.  Define the set $$\{X \,\,|\,\,  \text{trace}(A_i^T X A_i) \le 0, \,\, \forall i \}.$$
Is this set of matrices convex?

If the given set is convex, it would mean that the condition 
$$\text{trace}(A_i^T (\theta X_1 + (1 - \theta) X_2) A_i) \le 0$$
must be met.

$$\text{trace}(A_i^T (\theta X_1 + (1 - \theta) X_2) A_i)= \text{trace}(\theta A_i^T X_1 A_i + (1 - \theta)A_i^T X_2 A_i)$$
$$= \theta \, \text{trace}\big(A_i^T X_1 A_i\big) + (1 - \theta) \, \text{trace}\big(A_i^T X_2 A_i\big)$$
Since $0 < \theta < 1$, $\text{trace}\big(A_i^T X_1 A_i\big) \le 0$ and $(1 - \theta) \, \text{trace}\big(A_i^T X_2 A_i\big) \le 0$, the above value is always $\le 0$.

It can also be seen from the fact that the expression $\text{trace}(A_i^T X A_i)$ is the linear mapping of a linear mapping, and hence the resulting set is a convex set.

### (e)  Let $\|\cdot\|$ be the vector 2-norm.  Is the set $C = \{(x,y)\, |\,\, \|x\| \le  y \}$ convex?  Is it a cone?  Why?  What about the set $C = \{(x,y)\, |\,\, \|x\|^2 \le  y \}?$  Is this convex? Is this a  cone?  Here, $x$ is a vector and $y$ is a scalar.

**Part-1:**

Consider an $x_3$ such that $x_3 = \theta x_1 + (1-\theta) x_2$.
$$ \| \theta x_1 + (1-\theta) x_2 \| \le \theta \,\|x_1\| + (1-\theta) \,\|x_2\|$$

We know that $\|x_1\| \le y$ and $\|x_2\| \le y$. Therefore,
$$ \theta \,\|x_1\| + (1-\theta) \,\|x_2\| \le \theta y + (1-\theta)y = y. $$

Therefore, the set $C = \{(x,y)\, |\,\, \|x\| \le  y \}$ is convex.

To prove that it is a convex cone, the set should satisfy the condition $x \in C \Rightarrow ax \in C$.
$$ \|ax\| \le \|a\|\|x\| \le ay. $$

Therefore, the given set is a cone.

**Part-2:**

Consider an $x_3$ such that $x_3 = \theta x_1 + (1-\theta) x_2$.
$$ \| \theta x_1 + (1-\theta) x_2 \|^2 \le \theta^2\|x_1\|^2 + (1-\theta)^2\|x_2\|^2 +  2\theta(1-\theta)\|x_1\|\|x_2\|$$

It is known that $\|x\|^2 \le y$ and hence, $\|x\| \le \sqrt{y}$. Plugging these into the above equation,
$$ \le \theta^2 y + (1+\theta^2-2\theta)y + 2(\theta - \theta^2)y = y.$$

Therefore, the set $C = \{(x,y)\, |\,\, \|x\|^2 \le  y \}$ is convex.

Again, to prove that it is a convex cone, the set should satisfy the condition $x \in C \Rightarrow ax \in C$.
$$ \|ax\|^2 \le \|a\|^2\|x\|^2 \le a^2 y  $$

Therefore, the given set is not a cone.

### (a) Suppose the function $f(x,y)$ is convex.  Prove that 
  $$g(x) = \min_y f(x,y)$$
  is also convex, where $x,y\in \mathbb{R}^n.$


$\min_y f(x,y)$ implies that we would be minimizing $f$ over all $y$ and since $f(x, y)$ is a convex function in $y$, there is a unique minimum. Consider that minimum to be $f(x, y^*)$. 

Now, let us vary $f$ in $x$ - consider points $x_1$ and $x_2$ in $x$.
$$ f(\theta x_1 + (1-\theta)x_2, y^*) \le \theta f(x_1, y^*) + (1 - \theta) f(x_2, y^*) $$
Considering $ f(\theta x_1 + (1-\theta)x_2, y^*) = g(\theta x_1 + (1-\theta)x_2), f(x_1, y^*) = g(x_1) $ and $ f(x_2, y^*) = g(x_2) $, we have
$$  g(\theta x_1 + (1-\theta)x_2) \le \theta g(x_1) + (1 - \theta) g(x_2) $$

which is the Jensen's inequality and hence, the function $g(x)$ is also convex.

### (b)  Suppose $f$ is strongly convex.  Show that the gradient of $f$ is strongly monotone, i.e., there is a constant $m>0$ with 
  $$ \langle x-y,\nabla f(x) - \nabla f(y) \rangle \ge m\|x-y\|^2.$$


It is known that the function $f$ is strongly convex, which means that 
$$ f(y) \ge f(x) + (y-x)^{T} \nabla f(x) + \frac{m}{2} \|y - x\|^2. $$

Using this result, we can form the following two equations:
$$ f(x) \ge f(y) + (x-y)^{T} \nabla f(y) + \frac{m}{2} \|x - y\|^2, $$
$$ f(y) \ge f(x) + (y-x)^{T} \nabla f(x) + \frac{m}{2} \|y - x\|^2 $$

Combining the above equations,
$$ f(x) + f(y) \ge f(x) + f(y) + (x-y)^{T} \nabla f(y) - (x-y)^{T} \nabla f(x) + m \|y - x\|^2 $$

$$ \Rightarrow 0 \ge  -(x-y)^{T} (\nabla f(x) - \nabla f(y)) + m \|y - x\|^2 $$
$$ \Rightarrow (x-y)^{T} (\nabla f(x) - \nabla f(y)) \ge m \|y - x\|^2 $$

Hence,
$$ \langle x-y,\nabla f(x) - \nabla f(y) \rangle \ge m\|x-y\|^2. $$

# Problem 3
 Suppose that $g(x)$ is convex and $h(x)$ is concave (i.e. $-h(x)$ is convex).  Suppose we restrict both functions into a closed, convex set $C$ such that both $g(x)$ and $h(x)$ are always positive when $x \in C.$  Prove that the function
  $f(x) = g(x)/h(x)$ 
  is quasi-convex.  (note: It follows that every local minimum is also global)



A function $f : \textbf{R}^n \rightarrow \textbf{R}$ is called quasi-convex if its domain and all its sublevel sets 
$$ S_{\alpha} = \{x \in \textbf{dom}\ f\ |\ f(x) \leq \alpha \} $$
for $\alpha \in \textbf{R}$ are convex. 

Here we have $g(x) > 0$, $h(x) > 0$ for $x \in \textbf{C}$. Consider the following sublevel set:
$$ S_{\alpha} = \{x \in \textbf{C}\ |\ f(x) \leq \alpha \} $$
$$ \frac{g(x)}{h(x)} \le \alpha \Rightarrow g(x) - \alpha h(x) \le 0 $$

Since it is given that $g(x)$ is convex and $h(x)$ is concave (which means $-h(x)$ is convex), the above expression is convex. 

Since the sublevel set $S_{\alpha}$ is convex, we can conclude that $f(x)$ is quasi-convex.
